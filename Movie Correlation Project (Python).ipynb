```python
#Correlation in Python 

# First let's import the packages we will use in this project

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjust the configuration of the plots we will create

# pd.options.mode.chained_assignment = None



# Now we need to read in the data
df = pd.read_csv(r'C:\Users\jonat\Desktop\movies.csv')
```

```python
# Let's look at the data

df.head()
```

```python
# Checking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for our columns

df.dtypes
```

```python
df = df.dropna()
```

```python
# Change data type for column

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# see all data

pd.set_option('display.max_rows', None)
```

```python
# Drop duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
# df.drop_duplicates()
```

```python
df.head()
```

```python
# Scatter plot with budget vs gross rev.

plt.scatter(x = df['budget'], y = df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for film')


plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn

#this shows a positive correlation

sns.regplot(x='budget' , y='gross' , data=df , scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Start Correlation Analysis
```

```python
df.corr(method='pearson')  

#pearson, Kendall, spearson
```

```python
# High correlation between budget and gross
```

```python
#Visualize Correlation Matrix

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()

#notice brighter colors have higher correlation
```

```python
#looks at Company.
# The variable company is not numeric but we can create a numeric representation of it

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()        
```

```python
df.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
#organize to see higest correlation quickly using unstacking

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs

#
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
